## Iris Data Set Exploratory data analysis

Importing the Necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df['Species'].unique()
```

```python
df.drop(columns = ['Id'],axis=1, inplace = True)
```

```python
a=df.Species.str.split('Iris-',expand=True)
df['Species']=pd.DataFrame(a[1])
df.head()
```

```python
df['Species'].unique()
```

## Univariate Analysis

```python
df_setosa = df.loc[df['Species']=='setosa']
df_virginica = df.loc[df['Species']=='virginica']
df_versicolor = df.loc[df['Species']=='versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa),'o')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor),'o',color='green')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue='Species',size = 5)
```
